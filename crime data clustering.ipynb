# K Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
#Building elbow graph by using wcss to decide the k-value
#wcss-within cluster sum of squares
wcss = []
for i in range (1,11):
  kmeans = KMeans(n_clusters=i,random_state=0)
  kmeans.fit(crime_norm)
  wcss.append(kmeans.inertia_)
```

```python
#Plotting elbow graph
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('No of clusters')
plt.ylabel("wcss")
plt.show()
```

```python
clusters = KMeans(n_clusters=4)
clusters.fit(crime_norm)
clusters.labels_
```

```python
#creating clusterid 
crime['kcluster'] = clusters.labels_
```

```python
crime1=crime.drop('cluster',axis = 1)
```

```python
crime1
```

```python
crime1.iloc[:,1:5].groupby(crime1.kcluster).mean()
```

# DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime2=crime.iloc[:,1:5]
```

```python
crime2.values
```

```python
#standardization 
stscaler = StandardScaler().fit(crime2.values)
x = stscaler.transform(crime2.values)
```

```python
x
```

```python
dbscan = DBSCAN(eps = 2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['dcluster'])
```

```python

cl
```

```python

pd.concat([crime2,cl],axis=1)
```
